```python
import sys
import json
import re
from tabulate import tabulate
```

```python
profile_file = "memcached_syscalls.profile"
```

```python
with open(profile_file) as f:
    list_stats = f.read().split(',')
    fields = ["count", "native", "virtual"]
    feats = [dict(zip(fields, [int(x) for x in re.search("(\d*):(\d*)/(\d*)", elem).groups()])) for elem in list_stats]
    with open("syscall_list.txt") as g:
        syscall_list = g.read().split('\n')
        final_dict = dict(zip(syscall_list, feats))
        for k, v in final_dict.items():
            final_dict[k]["overhead"] = ((v["virtual"] / v["native"])-1)*100 if v["native"] else 0
```

```python
#for k, v in final_dict.items():
print ("{:<20} {:<10} {:<15} {:<15} {:<10}".format('Syscall', 'Count', 'Native (ns)', 'Virtual (ns)', 'Overhead (%)'))
for k, v in sorted(final_dict.items(), key=lambda x: x[1]["overhead"], reverse=True):
    print ("{:<20} {:<10} {:<15} {:<15} {:<10}".format(k, v["count"], v["native"], v["virtual"], v["overhead"]))
```
